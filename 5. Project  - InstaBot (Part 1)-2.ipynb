```python
followers = driver.find_elements_by_class_name('Y8-fY')
followers[1].click()
```

```python
import time
f  = driver.find_element_by_xpath("//div[@class='isgrP']")
followers_1 = []
scroll = 0
while scroll < 15:
    follower = driver.find_elements_by_xpath("//*[contains(@class, 'd7ByH')]")
    for i in follower:
        followers_1.append(i.text)
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', f)
    time.sleep(2)
    scroll += 1
```

```python
followers_foodtalkindia = followers_1[:500]
print(len(followers_foodtalkindia))
```

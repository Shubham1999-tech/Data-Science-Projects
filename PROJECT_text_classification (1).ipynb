# IMPLEMENTATION

```python
x_train=x_train.reset_index(drop=True)
x_test=x_test.reset_index(drop=True)
```

```python
#features
features=x_train.columns
```

```python
#convert training and testing into numpy array for better calculation
x_train=np.array(x_train)
x_test=np.array(x_test)
```

```python
x_train=x_train[:,1:]
x_test=x_test[:,1:]
```

```python
#fit function 
def fit(x_train,y_train,features):
    dictionary={}
    dictionary['total_data']=len(y_train)
    x1=list(set(y_train))
    for i in x1:
        dictionary[i]={}
        new=np.where(y_train==i)
        x_train_current=x_train[new]
        for j in range(len(features)-1):
            dictionary[i][features[j]]=x_train_current[:,j].sum()
        p=list(dictionary[i].values())
        dictionary[i]['total_words']=sum(p)-len(x_train_current)  
        dictionary[i]['class_p']=len(x_train_current)
    return dictionary               
```

```python
fit_dict=fit(x_train,y_train,features)
```

```python
#CALCULATE PROBABILITY FOR EVERY CLASS AND PREDICTSINGLEPOINT COMPARES THE ACCURACY FOR EVERY DOCUMENT AND PICK THE CLASS WITH MAXIMUM PROBABILITY
def probability(dictionary,x,current_class,features):
    output=np.log(dictionary[current_class]['class_p'])-np.log(dictionary['total_data'])
    for i in range(len(features)-1):
        classi=dictionary[current_class]['total_words']+len(features)-2
        if features[i]=='total_words' or features[i]=='class_p' or x[i]==0:
            continue   
        pwi=dictionary[current_class][features[i]]+1
        output+=(np.log(pwi)-np.log(classi))
    return output    
```

```python
###CALCULATE BEST CLASS FOR EVERY SINGLE DOCUMENT IN TEST DATA

def predictsingleword(fit_dict,x,features):
    classes=fit_dict.keys()
    best_p=-10000
    best_class=-1
    first_run=True
    for i in classes:
        if i=='total_data':
            continue
        p_current_class=probability(fit_dict,x,i,features)
        if (first_run or p_current_class>best_p):
            best_p=p_current_class
            best_class=i
            
        first_run=False
    return best_class    
```

```python
#PREDICT CLASS FOR EVERY DOCUMENT WITH THE HELP OF predictsinglepoint and probability FUNCTIONS
import re
def predict(fit_dict,x_test,features):
    y_pred=[]
    for i in range(len(x_test)):
        pred_class=predictsingleword(fit_dict,x_test[i],features)
        y_pred.append(pred_class)
    return y_pred     
```

```python
###PREDICTED ARRAY FOR TEST DATA
ans=predict(fit_dict,x_test,features)
```

```python
###CALCULATE SCORE OF ACTUAL TEST DATA AND PREDICTED TEST DATA
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ans)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,ans))
print(confusion_matrix(y_test,ans))
```
